```python
import pandas as pd 
import numpy as np

df = pd.read_csv('./data/dataset.csv')
```

```python
print(df.shape)
df.info()
df.describe()
```

```python
## Splitting the Dataset

from sklearn.model_selection import train_test_split

y_target = "GROSS_DEMAND_MW"  # set the target variable

df.drop(columns=[
    'TOTAL_DEMAND',
    'ROOFTOP_SOLAR_MW'
], inplace=True, errors='ignore')

X_train, X_temp = train_test_split(df, test_size=0.4, random_state=42)
X_val, X_test = train_test_split(X_temp, test_size=0.5, random_state=42)

y_train, y_val, y_test = (
    X_train.pop(y_target),
    X_val.pop(y_target),
    X_test.pop(y_target),
)

df_names = ['X_train', 'X_val', 'X_test']
```

```python
print(X_train.shape)
print(X_val.shape)
print(X_test.shape)
```

## Feature Engineering

## Data Transformation

### One-hot Encoding (OHE)

```python
from sklearn.preprocessing import OneHotEncoder

ohe_encoder = OneHotEncoder()

for name in df_names:
    df_ = globals()[name]

    globals()[name] = df_.copy()
```

### Feature Scaling

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
numeric_cols = X_train.select_dtypes(include=['number']).columns

scaler.fit(X_train[numeric_cols])

for name in df_names:
    df_ = globals()[name]
    
    df_[numeric_cols] = scaler.transform(df_[numeric_cols])

    for col in df_.columns:
        if df_[col].dtype == 'bool':
            df_[col] = df_[col].astype('category').cat.codes    

    globals()[name] = df_.copy()
```

```python
X_train.info()
```

### Target value log transform

```python
# Convert to datetime
for df_ in [X_train, X_val, X_test]:
    df_['DATE_TIME'] = pd.to_datetime(df_['DATE_TIME'])

    # Extract features
    df_['hour'] = df_['DATE_TIME'].dt.hour
    df_['day'] = df_['DATE_TIME'].dt.day
    df_['month'] = df_['DATE_TIME'].dt.month
    df_['dayofweek'] = df_['DATE_TIME'].dt.dayofweek

    # Drop original column
    df_.drop(columns=['DATE_TIME'], inplace=True)
```

## Modelling

```python
from sklearn.linear_model import LinearRegression

lin_reg = LinearRegression(
    fit_intercept=True
)

lin_reg.fit(X_train, y_train)

lin_train_preds = lin_reg.predict(X_train)
lin_val_preds = lin_reg.predict(X_val)
lin_test_preds = lin_reg.predict(X_test)
```

```python

import pandas as pd
import numpy as np


drop_col = 'DATE_TIME'

X_tr = X_train.drop(columns=[drop_col], errors='ignore')
X_v  = X_val.drop(columns=[drop_col],   errors='ignore')
X_te = X_test.drop(columns=[drop_col],  errors='ignore')


X_tr = X_tr.astype(float)
X_v  = X_v.astype(float)
X_te = X_te.astype(float)


print(f"X_tr shape : {X_tr.shape}")
print(f"X_v  shape : {X_v.shape}")
print(f"X_te shape : {X_te.shape}")
print(f"Columns    : {list(X_tr.columns)}")
```

```python
from sklearn.linear_model import Ridge

ridge_model = Ridge(alpha=1.0)
ridge_model.fit(X_tr, y_train)

ridge_train_preds = ridge_model.predict(X_tr)
ridge_val_preds   = ridge_model.predict(X_v)
ridge_test_preds  = ridge_model.predict(X_te)
```

```python
from sklearn.linear_model import LinearRegression

lin_model = LinearRegression(fit_intercept=True)
lin_model.fit(X_tr, y_train)

lin_train_preds = lin_model.predict(X_tr)
lin_val_preds   = lin_model.predict(X_v)
lin_test_preds  = lin_model.predict(X_te)
```

```python

from sklearn.ensemble import RandomForestRegressor

rf_model = RandomForestRegressor(
    n_estimators=300,
    max_depth=12,
    min_samples_leaf=4,
    n_jobs=-1,
    random_state=42
)
rf_model.fit(X_tr, y_train)

rf_train_preds = rf_model.predict(X_tr)
rf_val_preds   = rf_model.predict(X_v)
rf_test_preds  = rf_model.predict(X_te)

```

```python
from xgboost import XGBRegressor

xgb_model = XGBRegressor(
    n_estimators=1000,
    learning_rate=0.05,
    max_depth=6,
    random_state=42
)

xgb_model.fit(
    X_train, y_train,
    eval_set=[(X_val, y_val)],
    verbose=False
)

xgb_train_preds = xgb_model.predict(X_train)
xgb_val_preds = xgb_model.predict(X_val)
xgb_test_preds = xgb_model.predict(X_test)
```

```python
from catboost import CatBoostRegressor

cat_model = CatBoostRegressor(
    iterations=1000,
    learning_rate=0.05,
    depth=6,
    loss_function='RMSE',
    random_state=42
)

cat_model.fit(
    X_train, y_train,
    eval_set=(X_val, y_val),
    early_stopping_rounds=50,
    verbose=False
)

cat_train_preds = cat_model.predict(X_train)
cat_val_preds = cat_model.predict(X_val)
cat_test_preds = cat_model.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import pandas as pd

def get_metrics(y_true, y_pred, name):
    mse = mean_squared_error(y_true, y_pred)
    rmse = np.sqrt(mse)
    mae = mean_absolute_error(y_true, y_pred)
    r2 = r2_score(y_true, y_pred)
    
    return {
        'Model': name,
        'RMSE (MW)': round(rmse, 2),
        'MAE (MW)': round(mae, 2),
        'R2 Score': round(r2, 4)
    }


results = [
    get_metrics(y_test, lin_test_preds, "XGBoost"),
    get_metrics(y_test, xgb_test_preds, "XGBoost"),
    get_metrics(y_test, cat_test_preds, "CatBoost")
]

metrics_df = pd.DataFrame(results)
display(metrics_df)
```

```python
from sklearn.neural_network import MLPRegressor

mlp_model = MLPRegressor(
    hidden_layer_sizes=(128, 64, 32),  # 3 hidden layers
    activation='relu',
    solver='adam',
    learning_rate_init=0.001,
    max_iter=300,
    early_stopping=True,          # uses 10% of train as internal val
    validation_fraction=0.1,
    n_iter_no_change=20,          # stops if no improvement for 20 epochs
    random_state=42,
    verbose=False
)

mlp_model.fit(X_tr, y_train)

mlp_train_preds = mlp_model.predict(X_tr)
mlp_val_preds   = mlp_model.predict(X_v)
mlp_test_preds  = mlp_model.predict(X_te)
```

```python


from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np
import pandas as pd

def get_metrics(y_true, y_pred, model_name, split):
    rmse = np.sqrt(mean_squared_error(y_true, y_pred))
    mae  = mean_absolute_error(y_true, y_pred)
    r2   = r2_score(y_true, y_pred)
    return {
        'Model':     model_name,
        'Split':     split,
        'RMSE (MW)': round(rmse, 2),
        'MAE (MW)':  round(mae, 2),
        'R² Score':  round(r2, 4)
    }

results = []

models = {
    'Linear Regression': (lin_train_preds,  lin_val_preds,  lin_test_preds),   # fixed label
    'XGBoost':           (xgb_train_preds,  xgb_val_preds,  xgb_test_preds),
    'CatBoost':          (cat_train_preds,  cat_val_preds,  cat_test_preds),
    'MLP Neural Net':    (mlp_train_preds,  mlp_val_preds,  mlp_test_preds),
}

for name, (tr_p, v_p, te_p) in models.items():
    results.append(get_metrics(y_train, tr_p, name, 'Train'))
    results.append(get_metrics(y_val,   v_p,  name, 'Val'))
    results.append(get_metrics(y_test,  te_p, name, 'Test'))

metrics_df = pd.DataFrame(results)

# Pivot — clean wide table
pivot = metrics_df.pivot(index='Model', columns='Split', values=['RMSE (MW)', 'MAE (MW)', 'R² Score'])
pivot = pivot.reindex(columns=pd.MultiIndex.from_product(
    [['RMSE (MW)', 'MAE (MW)', 'R² Score'], ['Train', 'Val', 'Test']]
))
display(pivot.sort_values(('RMSE (MW)', 'Test')))
```
